## Defining files directory and importing files

```
#The input directory in the google dirve
input_data_path="/gdrive/My Drive/Colab Notebooks/Input_files/Original"
output_data_path="/gdrive/My Drive/Colab Notebooks/Input_files/produced"
os.listdir(input_data_path)
```

```
#the aisle of  items
aisles=pd.read_csv(os.path.join(input_data_path,'aisles.csv'))
print(aisles.head())
print(aisles.describe())
print(aisles.info())
```

```
#Department id
departments=pd.read_csv(os.path.join(input_data_path,'departments.csv'))
print(departments.head())
print(departments.describe())
print(departments.info())
```

```
#orders : products, number of item, reorder
order_products_prior=pd.read_csv(os.path.join(input_data_path,'order_products__prior.csv'))
print(order_products_prior.head())
print(order_products_prior.describe())
print(order_products_prior.info())
```

```
order_products_train=pd.read_csv(os.path.join(input_data_path,'order_products__train.csv'))
print(order_products_train.head())
print(order_products_train.describe())
print(order_products_train.info())
```

```
#Customers order: user id, order number, hour, days between orderes
orders=pd.read_csv(os.path.join(input_data_path,'orders.csv'))
print(orders.head())
print(orders.describe())
print(orders.info())
```

```
orders.eval_set.unique()
```

In the order dataset there are 3 different gatheogries. These entries were used to determine the Train, test and prior datasets in the original Kaggle competition.

```
#products: name, aisle, department
products=pd.read_csv(os.path.join(input_data_path,'products.csv'))
print(products.head())
print(products.describe())
print(products.info())
```

##Combining datasets

```
order_products_train.head()
```

```
order_products_prior.head()
```

```
#Combine order_products data set together
#every order includes which products
#there is no users information yet
order_product=pd.concat([order_products_prior,order_products_train]).sort_values(by=['order_id','product_id'])
order_product.head(20)
```

```
order_product.shape
```

```
products.shape
```

We add aisle_id and department_id for every products in the orders, but we don't need product, department or aisle names

```
order_product_info=order_product.join(products.set_index('product_id'),on='product_id').drop(columns=['product_name'])
print(order_product_info.shape)
print(order_product_info.describe())
order_product_info.head()
```

```
order_product_info=order_product_info.reset_index()
order_product_info=order_product_info.drop(columns=['index'])
order_product_info.head()
```

```
#to release some memory
del products,order_product,order_products_prior,order_products_train,departments,aisles
```

```
df=orders.join(order_product_info.set_index('order_id'),on='order_id')
#df.to_csv(os.path.join(input_data_path,'data_InstaCart.csv'), encoding='utf-8',index=False)
```

```
df.head(50)
```

## Exporting the final data into a cvs file

```
df.to_csv(os.path.join(output_data_path,'data_InstaCart_copy.csv'), encoding='utf-8',index=False)
```
